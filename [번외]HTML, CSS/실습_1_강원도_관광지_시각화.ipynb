```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.gangwon.to/gwtour/gangwondo_trip/tourist_spot/nature"
```

```python
response_text = requests.get(url)
response_text
```

```python
# html text 가져오기
html_text = response_text.content

soup = BeautifulSoup(html_text, 'html.parser')
soup
```

* `find` : 1개
* `find_all` : 여러 개
* `select` : 여러 개
* `select_one` : 1개


```python
sample_li = soup.select_one("#contents > div.contsArea.skinContainer > ul.nlist01.clearfix > li")
sample_li
```

```python
sample_text_div = sample_li.select_one("div.text")
sample_text_div
```

```python
category = sample_text_div.select_one("span").text
title    = sample_text_div.select_one("strong").text
desc     = sample_text_div.select_one("p").text

print(category, title, desc)
```

# 전페 페이지 크롤링

```python
last_page = 102
```

```python
url = "https://www.gangwon.to/gwtour/gangwondo_trip/tourist_spot/nature"
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
from tqdm import tqdm_notebook

datas = []

# 페이지 돌기
for page in tqdm_notebook(range(1, last_page + 1)):
  # 1. 서버에 전송할 데이터 준비
    # 이렇게만 넘기는 방식을 form-data
    # json으로 넘겨야 하면.. content-type: json이면
    # json.dumps(data) 감싸서 전송
  data = {"pageIndex": page}

  response = requests.post(url, data).content

  soup = BeautifulSoup(response, 'html.parser')
  total_ul = soup.select_one("#contents > div.contsArea.skinContainer > ul.nlist01.clearfix")
  all_li   = total_ul.select("li") # li가 여러 개 있으니까 select

  for li in all_li:
    text_div = li.select_one("div.text")

    category = text_div.select_one("span").text
    title    = text_div.select_one("strong").text
    desc     = text_div.select_one("p").text

    scrap_data = {
        "카테고리": category,
        "관광지명": title,
        "관광지소개": desc
    }

    print(scrap_data)

    datas.append(scrap_data)
```

```python
tour_df = pd.DataFrame(datas)
tour_df.head()
```
