# 결측값 채우기
* `fillna` : `nan`값을 특정 값으로 채워주는 함수


```python
df['키']
```

```python
df['키'].fillna(0) # '키' 컬럼에서 NaN인 데이터를 0으로 채우겠다.
```

```python
# 데이터 백업
df2 = df.copy()
df2.head()
```

복사된 `df2`의 `info()` 확인

```python
df2.info()
```

```python
# df2의 '키'의 NaN값을 0으로 채우고 df2에 적용
df2['키'] = df2['키'].fillna(0)
df2.info()
```

키를 0이 아닌 `평균`으로 채워보자

```python
df3 = df.copy()
df3.info()
```

```python
# 키의 평균값 구하기
mean_height = df['키'].mean()
mean_height
```

```python
df3['키'] = df3['키'].fillna(mean_height)
df3
```

[번외] 남녀 따로따로 평균 키 구하기

```python
df4 = df.copy()
girl_mean_height = df4.loc[df4['성별'] == '여자', '키'].mean()
girl_mean_height
```

```python
changed = df4.loc[df4['성별'] == '여자']['키'].fillna(girl_mean_height)
df4.loc[df4['성별'] == '여자', '키'] = changed
```

```python
df4
```

# 결측치가 있는 행을 제거
* `dropna()`

```python
df
```

dropna() 적용

```python
df.dropna()
```

dropna는 단순하게 nan값이 들어있으면 그 행을 삭제

## dropna의 axis
* `axis=0` : 행 삭제
* `axis=1` : 열 삭제

```python
df.dropna(axis=0) # NaN값이 한개라도 들어있는 행을 제거
```

```python
df.dropna(axis=1) # NaN값이 존재하는 열을 삭제
```

## dropna의 how 옵션
* `how='any'` : NaN이 한 개라도 있으면 제거 - 기본
* `how='all'` : 데이터가 모두 NaN인 경우에만 drop

```python
df.dropna(axis=0, how='all')
```

```python
import numpy as np
df.iloc[10] = np.nan
df
```

```python
df.dropna(axis=0, how='all')
```

# 중복 데이터 제거하기

```python
df['키']
```

```python
df['키'].drop_duplicates()
```

keep 옵션을 사용해서 유지하고 싶은 데이터를 선택할 수 있다.
* `keep='first'` : 첫 번째 데이터를 남긴다.
* `keep='last'` : 제일 마지막 데이터를 남긴다.

```python
df['키'].drop_duplicates(keep='first')
```

```python
df['키'].drop_duplicates(keep='last')
```

## 중복되는 행을 제거하기
* `keep`옵션을 사용할 수 있다.

```python
df.drop_duplicates("그룹") # 그룹 컬럼에서 중복된 행을 제거. 그룹이 중복되어 있으면 제거
```

```python
df.drop_duplicates("그룹", keep='last')
```

# 행 / 열 삭제

## 열 삭제
* `drop` 함수를 활용하며 `axis=1`로 주면 삭제가 됩니다.

```python
df.drop("그룹", axis=1)
```

여러 개의 컬럼도 동시에 제거가 가능

```python
df.drop(['그룹', '소속사'], axis=1)
```

## 행 삭제
* 컬럼을 제거 할 때는 컬럼의 이름을 사용
* 열 삭제 할 때는 index를 사용해서 삭제.
* `axis=0`으로 설정해야 한다.

```python
df.drop(3, axis=0)
```

```python
df.drop(1, axis=0)
```

```python
df.drop([1, 3, 5], axis=0)
```
