# CSV 파일 읽어오기 - 일반 파일
**CSV 파일이란?**
* Comma Sperated Value의 약자로써, 데이터가 쉼표로 구분 되어 있다.
* 엑셀을 로딩할 수도 있지만, 엑셀은 데이터와 관련이 없는 다양한 정보도 포함
  - 표의 굵기, 셀의 색상, 계산식 등등...
* CSV를 이용하면 순수하게 데이터만 가져올 수 있다. 엑셀보다 속도가 훨씬 빠르다.
```csv
  이름,나이,키
  소민호,34,180.1
  김땡땡,33,175.5
```

````python
# read_csv를 활용하면 데이터를 손쉽게 가져올 수 있다.
pd.read_csv("korean-idol.csv")
````

영어권이 아닌 데이터를 그냥 read_csv로 불러오면 글자가 깨지는 경우가 종종 있다.
* ?????, 외계어가 나오는 경우(알 수 없는 특수기호)
* `read_csv("파일 이름", encoding='utf-8')` 또는 `read_csv("파일 이름", encoding='euc-kr')`
