# 데이터 정렬하기

## 오름차순 인덱스 정렬하기

```python
df.sort_index() # 오름차순 정렬이 기본 ascending=True
```

## 내림차순 인덱스 정렬하기
* `ascending=False` 옵션을 사용하면 내림차순 정렬

```python
df.sort_index(ascending=False)
```

## 컬럼 오름차순 정렬
* `by` 옵션을 이용해서 정렬의 기준이되는 컬럼을 지정. 2개 이상도 가능
* `ascending` 옵션을 이용해서 오름 / 내림 차순 설정 가능

```python
df.sort_values(by="키")
```

## 컬럼 내림차순 정렬

```python
df.sort_values(by="키", ascending=False)
```

# 두 개 이상의 컬럼을 사용해서 정렬
* 앞 쪽에 배치된 컬럼을 먼저 고려해서 정렬
* `by=['키', '브랜드평판지수']` -> 키를 먼저 고려해서 정렬하고, 브랜드평판지수로 정렬

```python
df.sort_values(by=["키", "브랜드평판지수"])
```

```python
df.sort_values(by=["키", "브랜드평판지수"], ascending=[False, True])
```
